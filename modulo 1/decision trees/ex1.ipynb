```python
import pandas as pd

arquivo = pd.read_csv('./Admission_Predict.csv')
```

```python
arquivo.head()
```

```python
arquivo.drop('Serial No.', axis=1, inplace=True)
```

```python
y = arquivo['Chance of Admit ']
x = arquivo.drop('Chance of Admit ', axis=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score, KFold

kfold = KFold(n_splits=5, shuffle=True, random_state=7)

modelo = DecisionTreeRegressor()
resultado = cross_val_score(modelo, x, y, cv=kfold)

print("Coeficiente de determinação R2: ", resultado.mean())
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

minimos_split = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
maximo_nivel = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
algoritmo = ['mse', 'friedman_mse', 'mae']
valores_grid = { 'min_samples_split': minimos_split, 'max_depth': maximo_nivel, 'criterion': algoritmo } 

modelo = DecisionTreeRegressor()

gridDecisionTree = GridSearchCV(estimator = modelo, param_grid = valores_grid, cv = 5, scoring='r2')
gridDecisionTree.fit(x, y)

print("Mínimo split: ", gridDecisionTree.best_estimator_.min_samples_split)
print("Máximo nível: ", gridDecisionTree.best_estimator_.max_depth)
print("Algoritmo: ", gridDecisionTree.best_estimator_.criterion)
print("Erro médio quadrático: ", gridDecisionTree.best_score_)
```
