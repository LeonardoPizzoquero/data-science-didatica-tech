```python
import pandas as pd

pd.set_option('display.max_columns', 64)
pd.set_option('display.max_rows', 64)

df_train = pd.read_csv('./Data_train_reduced.csv')
df_test = pd.read_csv('./test_data.csv')
```

```python
missing = df_train.isnull().sum()
missing_percent = (missing / df_train.shape[0]) * 100
print(missing_percent)
```

```python
df_train.drop(['q8.20', 'q8.18', 'q8.17', 'q8.8', 'q8.10', 'q8.9', 'q8.2', 'Respondent.ID', 'Product', 'q1_1.personal.opinion.of.this.Deodorant'], axis=1, inplace=True)
```

```python
df_train.drop(['Product.ID'], axis=1, inplace=True)
```

```python
df_train['q8.12'].fillna(df_train['q8.12'].median(), inplace=True)
df_train['q8.7'].fillna(df_train['q8.7'].median(), inplace=True)
```

```python
missing = df_train.isnull().sum()
missing_percent = (missing / df_train.shape[0]) * 100
print(missing_percent)
```

```python
y = df_train['Instant.Liking']
X = df_train.drop('Instant.Liking', axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold

stratified_kfold = StratifiedKFold(n_splits=15)

model = LogisticRegression(max_iter=1000)
result = cross_val_score(model, X, y, cv = stratified_kfold)

print(result.mean())
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

values_C = np.array([0.01, 0.1, 0.5, 1, 2, 3, 5, 10, 20, 50, 100])
penalty = ['l1', 'l2']
params_grid = {'C': values_C, 'penalty': penalty}

model = LogisticRegression(max_iter=2000, solver='liblinear')

grid_search = GridSearchCV(estimator=model, param_grid=params_grid, cv=5)
grid_search.fit(X, y)

print("Acurácia: %.3f" % (grid_search.best_score_ * 100))
print("Parâmetro C: %.2f" % (grid_search.best_estimator_.C))
print("Regularização: %s" % (grid_search.best_estimator_.penalty))
```
