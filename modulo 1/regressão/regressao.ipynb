```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# Gerando dados de regressão (200 amostras, 1 feature, 1 target, ruído de 30%)
x, y = make_regression(n_samples=200, n_features=1, noise=30)

print(x.shape, y.shape)

# Plotando os dados
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor

# Instanciando o modelo
modelo = LinearRegression()
modelo2 = GradientBoostingRegressor()
```

```python
modelo.fit(x, y)
modelo2.fit(x, y)
```

```python
modelo2.feature_importances_
```

```python
modelo.intercept_
```

```python
modelo.coef_
```

```python
import numpy as np
plt.scatter(x, y)
xreg = np.arange(-3, 3, 1)

# y = mx + b = coef*x + intercept = 74.15772825*x + 2.655103363536047
plt.plot(xreg, 74.15772825*xreg+2.655103363536047, color='red')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
modelo.fit(x_treino, y_treino)
```

```python
resultado = modelo.score(x_teste, y_teste)
```

## Exercício de regressão com dataset King County Venda de casas

```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')

df.head()
```

```python
df.shape
```

```python
df = df.drop(['id', 'date', 'zipcode', 'lat', 'long'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

y = df['price']
x = df.drop('price', axis=1)

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
modelo.fit(x_treino, y_treino)
```

```python
modelo.score(x_teste, y_teste)
```
