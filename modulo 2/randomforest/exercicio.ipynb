```python
import pandas as pd

pd.set_option('display.max_columns', 81)

arquivo = pd.read_csv('BRAZIL_CITIES.csv', sep=';', decimal=",")
```

```python
arquivo.head()
```

```python
arquivo.drop(['CITY', 'IDHM Ranking 2010', 'IDHM_Renda', 'IDHM_Longevidade', 'IDHM_Educacao', 'LONG', 'LAT', 'GVA_MAIN', 'REGIAO_TUR', 'MUN_EXPENDIT', 'HOTELS', 'BEDS', 'Pr_Agencies', 'Pu_Agencies', 'Pr_Bank', 'Pu_Bank', 'Pr_Assets', 'Pu_Assets','UBER','MAC','WAL-MART'], axis=1, inplace=True)
```

```python
arquivo.head()
```

```python
arquivo['AREA']= arquivo['AREA'].str.replace(',', '')
```

```python
arquivo['CATEGORIA_TUR'].describe()
```

```python
arquivo['CATEGORIA_TUR'].value_counts()
```

```python
arquivo['CATEGORIA_TUR'].fillna('D', inplace=True)
```

```python
state_encode = pd.get_dummies(arquivo['STATE'], prefix='state')
rural_encode = pd.get_dummies(arquivo['RURAL_URBAN'], prefix='rural')
categoria_encode = pd.get_dummies(arquivo['CATEGORIA_TUR'], prefix='categoria')

arquivo.drop(['STATE', 'RURAL_URBAN', 'CATEGORIA_TUR'], axis=1, inplace=True)
```

```python
concat = [arquivo, state_encode, rural_encode, categoria_encode]
arquivo = pd.concat(concat, axis=1)
```

```python
arquivo.head()
```

```python
arquivo.isna().sum()
```

```python
arquivo = arquivo.dropna()
```

```python
X = arquivo.drop('IDHM', axis=1)
y = arquivo['IDHM']
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

scaler = MinMaxScaler()
X = scaler.fit_transform(X)

pca = PCA(n_components=15)
X_PCA = pca.fit_transform(X)
```

```python
print("Variância", pca.explained_variance_ratio_)
```

```python
sum(pca.explained_variance_ratio_)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score

kfold = KFold(n_splits=10, shuffle=True, random_state=42)
modelo = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=28064212)

resultado = cross_val_score(modelo, X_PCA, y, cv=kfold)

print("R2: %.3f (%.3f)" % (resultado.mean(), resultado.std()))
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV

minimos_split = np.array([2, 3, 4])
maximo_nivel = np.array([3, 5, 7, 9, 11, 14])
minimo_leaf = np.array([3, 4, 5, 6])
grid = {'min_samples_split': minimos_split, 'max_depth': maximo_nivel, 'min_samples_leaf': minimo_leaf}

modelo = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=28064212)

gridRandomForest = RandomizedSearchCV(estimator=modelo, param_distributions=grid, n_iter=100, cv=3, random_state=28064212, n_jobs=-1)

gridRandomForest.fit(X_PCA, y)

print("Mínimo split: ", gridRandomForest.best_estimator_.min_samples_split)
print("Máximo nível: ", gridRandomForest.best_estimator_.max_depth)
print("Mínimo leaf: ", gridRandomForest.best_estimator_.min_samples_leaf)
print("R2: %.3f" % (gridRandomForest.best_score_))
```
